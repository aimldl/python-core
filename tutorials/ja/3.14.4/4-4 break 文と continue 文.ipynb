## [4.4. break 文と continue 文](https://docs.python.org/ja/3.14/tutorial/controlflow.html#break-and-continue-statements)


```python
break 文は、その break 文を内包している最も内側にある for 文または while 文から抜け出すことができます:
```

```python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(f"{n} は {x} * {n//x} と等しい")
            break
```

```python
for n in range(2, 10):
    print(f"n={n}")
    for x in range(2, n):
        print(f"{n}, {x}")
        if n % x == 0:
            print(f"{n} は {x} * {n//x} と等しい")
            break
```

continue 文はループの次のイテレーションを実行します:

```python
for num in range(2, 10):
    if num % 2 == 0:
        print(f"{num}: 偶数を見つけた")
        continue
    print(f"{num}: 奇数を見つけた")
```
