### [4.9.7. ドキュメンテーション文字列](https://docs.python.org/ja/3.14/tutorial/controlflow.html#documentation-strings)
ドキュメンテーション文字列の内容と書式に関する慣習をいくつか挙げます。

最初の行は、常に対象物の目的を短く簡潔にまとめたものでなくてはなりません。簡潔に書くために、対象物の名前や型を明示する必要はありません。名前や型は他の方法でも得られるからです (名前がたまたま関数の演算内容を記述する動詞である場合は例外です)。最初の行はピリオドで終わっていなければなりません。

以下に複数行のドキュメンテーション文字列の例を示します:

```python
def my_function():
    """なにもしないがドキュメントは書く。

    本当になにもしない。

        >>> my_function()
        >>>
    """
    pass
```

ドキュメンテーション文字列中にさらに記述すべき行がある場合、二行目は空行にし、まとめの行と残りの記述部分を視覚的に分離します。つづく行は一つまたはそれ以上の段落で、対象物の呼び出し規約や副作用について記述します。

The Python parser strips indentation from multi-line string literals when they serve as module, class, or function docstrings.

```python
print(my_function.__doc__)
```
