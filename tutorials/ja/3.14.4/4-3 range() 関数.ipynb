## [4.3. range() 関数](https://docs.python.org/ja/3.14/tutorial/controlflow.html#the-range-function)

この関数は算術型の数列を生成します:

```python
for i in range(5):
    print(i)
```

* 数列にわたって反復を行う必要がある場合、組み込み関数 range() が便利です。
* 指定した終端値は生成されるシーケンスには入りません。

range(10) は 10 個の値を生成し、長さ 10 のシーケンスにおける各項目のインデクスとなります。

```python
for i in range(10):
    print(i)
```

range を別の数から開始したり、

```python
list( range(5, 10) )
```

他の増加量を指定することもできます:

```python
list( range(0, 10, 3) )
```

増加量は時に 'ステップ(step)' と呼ばれることもあります

負の値を生成することもできます

```python
list( range(-10, -100, -30) )
```

あるシーケンスにわたってインデクスで反復を行うには、 range() と len() を次のように組み合わせられます:

```python
a = ['Mary', 'had', 'a', 'little', 'lamb']
for i in range(len(a)):
    print(i, a[i])
```

しかし、多くの場合は enumerate() 関数を使う方が便利です。 ループのテクニック を参照してください。

range を直接出力すると変なことになります:

```python
range(10)
```

range() が返すオブジェクトはリストではありません。
* これは、イテレートした時に望んだ数列の連続した要素を返すオブジェクトです。
* しかし実際にリストを作るわけではないので、スペースの節約になります。

このようなオブジェクトは イテラブル と呼ばれます。 イテラブルを受け取る関数の例には sum() があります:

```python
sum( range(4) ) # 0 + 1 + 2 + 3
```
