## [4.7. match 文](https://docs.python.org/ja/3.14/tutorial/controlflow.html#match-statements)

* switch文と表面的には似てて...
* どのcaseにもマッチしない場合はどの分岐も実行されません。


match 文は1つの式を指定し、その値と次に続く1つ以上のcaseブロックに指定されたパターンを比較します。この機能はCやJava、JavaScript(や他の多数の言語)のswitch文と表面的には似ていますが、RustやHaskellのパターンマッチングにより似ています。最初にマッチしたパターンのみが実行され、コンポーネント(シーケンスの要素やオブジェクトの属性)から値を取り出して変数に代入することもできます。どのcaseにもマッチしない場合はどの分岐も実行されません。

最も単純な形式は、対象の値に対して1つ以上のリテラルです:

````python
import sys
print(sys.version)
````

````python
import sys
print(sys.version)
````

````python
def http_error(status):
    match status:
        case 400:
            return "Bad request"
        case 404:
            return "Not found"
        case 418:
            return "I'm a teapot"
        case _:
            return "Something's wrong with the internet"
````

최신 Python의 보안 정책(PEP 668)에 정확히 걸리셨네요!

Homebrew를 통해 설치한 최신 버전의 Python(3.11 이상)에서는 시스템 전체의 패키지 의존성이 꼬이는 것을 방지하기 위해, 사용자가 전역(Global) 환경에 pip install을 직접 하는 것을 기본적으로 차단하고 있습니다. 그래서 externally-managed-environment 에러가 발생한 것입니다.

이 문제를 해결하고 ipykernel을 설치하는 두 가지 방법이 있습니다. 상황에 맞게 선택해 보세요.

방법 1: 가장 빠른 해결책 (강제 설치)
에러 메시지 하단에도 나와 있듯이, --break-system-packages 플래그를 추가하여 이 보안 정책을 무시하고 강제로 설치하는 방법입니다. 단순히 Jupyter 커널(ipykernel)만 실행하려는 목적이라면 이 방법이 가장 빠릅니다.

터미널에 아래 명령어를 실행해 보세요.
```bash
/opt/homebrew/bin/python3 -m pip install ipykernel -U --user --force-reinstall --break-system-packages
```

설치가 완료되면, 이전 답변의 안내처럼 VS Code 창을 새로고침(Developer: Reload Window)하고 커널을 다시 연결해 주시면 됩니다.

방법 2: 가장 안전한 정석 (가상환경 사용)
앞으로 해당 코드에서 다양한 외부 패키지들을 추가로 설치하며 작업하실 계획이라면, 글로벌 환경을 건드리지 않고 프로젝트별로 가상환경(Virtual Environment)을 만들어 사용하는 것이 가장 권장되는 방식입니다.

VS Code 터미널(작업 중인 프로젝트 폴더 경로)에서 아래 명령어들을 순서대로 실행하세요.

1. 가상환경 생성 (이름을 .venv로 지정):
```bash
python3 -m venv .venv
```

2. 가상환경 활성화:
```bash
source .venv/bin/activate
```

(성공하면 터미널 프롬프트 맨 앞에 (.venv)라는 표시가 생깁니다.)

3. 가상환경 내부에 ipykernel 설치 (이때는 강제 플래그가 필요 없습니다):
```bash
pip install ipykernel
```
VS Code에 가상환경 커널 연결하기:
설치가 끝난 후, Jupyter Notebook 우측 상단의 커널 선택에서 Select Another Kernel... > Python Environments를 누르시면, 방금 생성한 Python 3.14.x ('.venv': venv) 항목이 새로 나타날 것입니다. 이를 선택해서 실행하시면 충돌 없이 완벽하게 작동합니다!


최근 Python 생태계에서 가장 주목받는 도구인 uv가 맞습니다.

uv는 파이썬 코드 포매터인 Ruff를 만든 Astral에서 Rust 언어로 개발한 초고속 파이썬 패키지 및 환경 관리자입니다. 기존의 venv와 pip를 완벽하게 대체하면서 패키지 설치 속도가 기존보다 수십 배 이상 빠르고, 환경이 꼬이는 문제도 아주 깔끔하게 해결해 줍니다.

uv를 사용해서 아주 빠르고 쾌적하게 가상환경을 세팅하고 ipykernel을 설치하는 방법을 단계별로 알려드릴게요.

 그러면 대신 MacOS의 Python 버전을 그냥 3.14로 변경하는 방법은 없어?

macOS 시스템 자체의 내장 Python 버전을 3.14로 완전히 덮어씌우거나 변경하는 것은 불가능하며, 절대 권장하지 않습니다.

* macOS 시스템 안정성 문제: macOS 내부의 여러 기본 프로그램과 Xcode 관련 도구들은 시스템에 기본 설치된 특정 버전(현재 3.9.6)의 Python에 의존하여 작동합니다. 이 버전을 강제로 최신 버전으로 바꾸거나 삭제하면 OS의 일부 기능이 망가질 수 있습니다.

*  애플의 시스템 무결성 보호 (SIP): 애플은 보안을 위해 사용자가 관리자 권한을 쓰더라도 시스템 핵심 폴더(/usr/bin/ 등)에 있는 기본 프로그램을 수정하거나 삭제하지 못하도록 시스템 단에서 강력하게 막아두었습니다.

그래서 PATH 환경 변수를 수정하는 방식이 전 세계 Mac 개발자들이 사용하는 가장 표준적이고 안전한 방법입니다.

macOS가 백그라운드에서 자기 일을 할 때는 내장된 3.9.6 버전을 쓰도록 내버려 둡니다.

대신, 사용자가 터미널을 열어 명령어를 치거나 VS Code를 실행할 때만 Homebrew로 설치한 3.14 버전을 "우선적으로(기본으로)" 찾도록 경로를 가로채는 방식입니다.
