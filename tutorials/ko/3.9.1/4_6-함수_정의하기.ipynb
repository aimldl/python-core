```
# 피보나치 수열을 임의의 한도까지 출력하는 함수
def fib(n):
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()
```

```
fib(2000)
```

```
# 피보나치 수열을 print 하는 대신, 리스트를 리턴하는 함수
def fib2(n):
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
    return result
```

```
f100 = fib2(100)
f100
```
