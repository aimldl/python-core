```
# 4.7.1. 기본 인자 값
#   https://docs.python.org/ko/3.9/tutorial/controlflow.html#default-argument-values
def ask_ok(prompt, retries = 4, reminder='Please try again!'):
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('Invalid user response')
        print(reminder)
```

```
# 이 함수는 하나의 필수 인자와 두 개의 선택 인자를 받아들입니다.
# 즉, 1개, 2개, 3개의 인자를 입력받을 수 있습니다.
# 1개
#   ask_ok('정말 끝내길 원하세요?')
# 2개
#   ask_ok('파일을 덮어써도 좋습니까?', 2)
# 3개
#   ask_ok('파일을 덮어써도 좋습니까?', 2, '자, 예나 아니요로만 답하세요!')
```

```
# review: 기본값은 한 번만 값이 구해집니다.
def f(a, L=[]):
    L.append(a)
    return L

# 다음 함수는 계속되는 호출로 전달된 인자들을 누적합니다.
print(f(1))
print(f(2))
print(f(3))
# 그러므로 연속된 호출 간에 기본값이 공유됩니다.
```

```
# 연속된 호출 간에 기본값이 공유하고 싶지 않으면
def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
```

```
print(f(1))
print(f(2))
print(f(3))
```
