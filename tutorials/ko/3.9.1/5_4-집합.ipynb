```
a = set('abracadabra')
a
```

```
b = set('alacazam')
b
```

```
a - b
```

```
a | b
```

```
a & b
```

```
# Letter in a or b, but not both
a ^ b
```

## 집합 컴프리헨션

```
a = {x for x in 'abracadabra' if x not in 'abc'}
a
```
