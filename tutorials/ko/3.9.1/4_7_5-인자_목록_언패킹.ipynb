```
# 4.7.5. 인자 목록 언 패킹
#   리스트, 튜플의 element들을 언패킹 해야할 때,
#     *연산자 혹은 **연산자
#   를 사용합니다.
#
#   함수 호출 시 분리된 위치 인자를 요구할 때가 있습니다.
#     예를 들어, range함수는 별도의 start, stop 인자를 기대합니다.
#   리스트, 튜플의 값을 넣을 때 언패킹하면 됩니다.
```

```
# review: args라는 리스트가 있을 때
args = [3, 6]

# range 함수는 별도의 start, stop 인자를 기대합니다.
# * 연산자를 사용해서 함수를 호출하면 됩니다.
list(range(*args))
```

```
# 아래와 동일합니다.
list(range(3, 6))
```

```
# review: ** 연산자를 써서 키워드 인자를 전달할 수 있습니다.

def parrot(voltage, state='a stiff', action='voom'):
    print("-- This parrot wouldn't", action, end=' ')
    print("if you put", voltage, "volts through it.", end=' ')
    print("E's", state, '!')
```

```
d = {'voltage': 'four million', 'state': "bleedin' demised", 'action': 'VOOM'}
parrot(**d)
```

```
# 아래와 동일합니다.
parrot('four million', "bleedin' demised", 'VOOM')
```
