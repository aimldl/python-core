## `raise`문
* `raise`문은 프로그래머가 지정한 예외를 강제로 
발생시킵니다.
* 'raise`에 발생시키고 싶은 예외를 인자에 넣습니다.
  * 예외 인스턴스
  * 예외 클래스
    * 예외 클래스는 [Exception](https://docs.python.org/ko/3.9/library/exceptions.html#Exception)을 상속하는 클래스입니다.

```
raise NameError('HIThere')
```

```
raise ValueError()
```

()를 생략할 수 있습니다

```
raise ValueError
```

```
raise ValueError('test')
```

만약 예외가 발생했는지 알아야 하지만 처리하고 싶지 않은 경우에
* raise문을 쓰면 그 예외를 다시 발생시킬 수 있습니다.

```
try:
    raise NameError('HiThere')
except NameError:
    print('An exception flew by!')
    raise
```

위의 경우
* 예외가 발생해서 `print`문의 내용이 출력되었습니다.
* 그리고 `---------------------------------------------------------------------------` 가 출력되었습니다.
* 그 밑부분은 `raise`로 인해 발생한 출력입니다.

`raise`가 없는 아래의 출력을 비교해보세요.

```
try:
    raise NameError('HiThere')
except NameError:
    print('An exception flew by!')
```
