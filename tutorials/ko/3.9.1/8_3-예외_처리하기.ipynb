````
while True:
    try:
        x = int(input('Please enter a number: '))
        break
    except ValueError:
        print('Oops! That was no valid number. Try again')
````

except절은 튜플로 여러 개의 예외를 지정할 수 있습니다.

```python
except (RuntimeError, TypeError, NameError):
    pass
```

except절에 클래스가 올 경우, 예외와 같은 클래스가 매치됩니다. 그리고 그 클래스의 베이크 클래스도 매치됩니다. 

````
class B(Exception):
    pass

class C(B):
    pass

class D(C):
    pass

for cls in [B, C, D]:  # cls = class
    try:
        raise cls()
    except D:
        print('D')
    except C:
        print('C')
    except B:
        print('B')
````

위의 예에서 except절의 순서가 반대가 되면 인쇄되는 결과가 바뀝니다.

````
class B(Exception):
    pass

class C(B):
    pass

class D(C):
    pass

for cls in [B, C, D]:  # cls = class
    try:
        raise cls()
    except B:
        print('B')
    except C:
        print('C')
    except D:
        print('D')
````

* 왜냐하면 except절에 나오는 클래스 혹은 베이스 클래스가 매치 되기 때문이다. 
* 클래스 C는 그 베이스 클래스인 B도 매치되므로 `except B:`에서 매치되어 버립니다.
* 마찬가지로 클래스 D는, C와 B에 매치되므로 `except B:`에서 매치됩니다.

except절은 예외 이름을 생략할 수 있습니다.
* 생략하면 와일드 카드 역할을 하므로, 모든 것에 매치됩니다.
* 그러므로 여러개의 except절의 마지막에서 `else절`과 유사한 기능을 합니다.

## `else절`의 사용 방법
* `try ... except`문에 `else`절이 쓰일 수 있습니다.
* `else`절은 항상 모든 `except`절의 마지막에 나와야 합니다.
* `try`절이 예외를 일으키지 않을 때, 실행되어야 하는 코드에 유용합니다.
* `try`절에 코드를 추가하는 것보다, `else`절에 추가하는 게 좋습니다.
* revisit: `try ... except`문에 의해 보호되고 있는 코드가 일으키지 않은 예외를 우연히 잡게 되는 것을 방지하기 때문입니다. (알듯 말듯한 표현)
  

````
import sys

for arg in sys.argv[1:]:
    try:
        f = open(arg, 'r')
    except OSError:
        print('cannot open', arg)
    else:
        print(arg, 'has', len(f.readlines()), 'lines')
        f.close()
````

## 예외의 인자 (exceptions' argument)
* 예외가 발생할 때 연관된 값을 가질 수 있습니다.
* 이런 값은 예외의 인자에 저장됩니다.
* 예외 타입 (exception type)에 따라 
  * 인자에 값이 있을 수도 없을 수도 있고
  * 인자의 타입도 달라집니다.



````
try:
    raise Exception('spam', 'eggs')
except Exception as inst:
    print(type(inst))  # the exception instance
    print(inst.args)   # arguments stored in .args
    print(inst)        # __str__ allows args to be printed directly.
                       # but may be overriden in exception subclasses
    x, y = inst.args   # unpack args
    print('x=', x)
    print('y=', y)
````

`try절`에 직접 등장하는 예외뿐만 아니라, (간적적으로) 호출되는 내부 함수에서 발생하는 예외도 처리합니다.

````
def this_fails():
    x = 1/0

try:
    this_fails()
except ZeroDivisionError as err:
    print('Handling run-time error:', err)
````
