# 5.2. `del`문
* https://docs.python.org/ko/3.9/tutorial/datastructures.html#the-del-statement

* 리스트에서 값 대신 인덱스를 사용해서 항목을 삭제할 때 쓰입니다.
* 변수를 삭제하는데 사용될 수 있습니다.
* `del`의 다른 용도를 뒤에서 보게 될 예정입니다.

```
a = [-1, 1, 66.25, 333, 333, 1234.5]
del a[0]
a
```

```
del a[2:4]
a
```

```
del a[:]
a
```

변수 자체를 삭제하는데 사용될 수 있습니다.

```
del a
```

이후에 a를 참조하면 에러가 발생합니다.
