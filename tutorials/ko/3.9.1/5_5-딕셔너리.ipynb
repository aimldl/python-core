## 키 인덱싱
* 숫자로 인덱싱되는 시퀀스 자료형과 달리
* 키 (key)로 인덱싱됩니다.

* 모든 불변형 (immutable)을 키로 사용할 수 있습니다.
  * 문자열
  * 숫자
  * 튜플
    * 튜플이 문자열, 숫자, 튜플들만 포함하면 키로 사용할 수 있습니다.

* 직/간접적으로 가변형 객체를 포함하면 키로 사용할 수 없습니다.
  * 리스트
    * 인덱스 대입, 슬라이드 대입
    * append(), extend() 같은 메서드들로
    * 값이 수정될 수 있기 때문입니다.

## 키: 값 쌍 (key: value pair)
딕셔너리는 키: 값 쌍 (key: value pair)의 집합으로 생각하면 좋습니다.

```
tel = {'jack': 4098, 'sape': 4139}
tel['guido'] = 4127
tel
```

```
tel['jack']
```

```
del tel['sape']
tel
```

```
tel['irv'] = 4127
tel
```

## `list(d)`
`list(d)`는 딕셔너리 `d`의 모든 키의 리스트를 삽입 순서대로 리턴합니다.

```
list(tel)
```

## `sorted(d)`
`sorted(d)`는 `d`의 모든 키의 리스트를 정렬해서 리턴합니다.

```
sorted(tel)
```

## `in` 키워드
하나의 키가 딕셔너리에 있는 지 검사하기 위해서 사용됩니다.

```
'guido' in tel
```

```
'sape' in tel
```

```
'jack' not in tel
```

## `dict()` 생성자
키-값 쌍들의 시퀀스에서 딕셔너리를 직접 구성합니다.

```
# key_value_pair_sequence
list_of_tuples = [('sape', 4139), ('guido', 4127), ('jack', 4098)]
type(list_of_tuples)
```

```
dict(list_of_tuples)
```

## 딕셔너리 컴프리헨션 (dictionary comprehension)
임의의 키와 값의 표현식에서 딕셔너리를 만들 수 있습니다.

```
# key = x, value = square of x (x**2)
{x: x**2 for x in (2, 4, 6)}
```

## 키워드 인자의 사용
키가 간단한 문자열일 때, 키워드 인자를 써서 간단하게 키:값 쌍을 지정할 수 있습니다.

```
dict(sape=4139, guido=4127, jack=4098)
```
