#  4.7.7. 도큐멘테이션 문자열 (Documentation strings, docstrings)
* [도큐멘테이션 문자열](https://docs.python.org/ko/3.9/tutorial/controlflow.html#documentation-strings)
* [4.7.7. Documentation Strings](https://docs.python.org/3.9/tutorial/controlflow.html#documentation-strings)


문서화를 할 때 형식에 대한 몇 가지 관례
* 첫 줄: 요약
  * 객체의 목적을 간결하게 요약합니다.
  * 간결함을 위해서 
    * 객체의 이름
    * 객체의 타입
  * 을 명시적으로 언급하지 않습니다.

여러 줄이 있을 경우
* 두 번째 줄: 없음
  * 첫 줄의 요약과 나머지 설명을 시각적으로 분리

  나머지 줄
  * 하나 혹은 그 이상의 문단으로 설명
    * 객체의 호출 규약
    * 부작용 등


```
# 예: multi-line docstring
def my_function():
    """ Do nothing, but document it.

    No, really, it doesn't do any thing.
    """
    pass
```

```
print(my_function.__doc__)
```
