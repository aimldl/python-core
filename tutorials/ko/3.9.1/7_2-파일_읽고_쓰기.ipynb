## 7.2.1. 파일 객체의 매소드
* https://docs.python.org/ko/3.9/tutorial/inputoutput.html#methods-of-file-objects

### f.read(size)
* size 만큼 데이터를 읽어옴
* 생략 시 전체를 읽어옴
* 파일 끝에 도달하면 빈 문자열 `"`을 리턴

````
with open('7-input.txt') as f:
    print(f.read())
````

### f.readline() 
* 파일에서 한 줄을 읽어옴.
* \n는 문자열의 끝에 보존됨

````
with open('7-input.txt') as f:
    print(f.readline())
    print(' ')
    print(f.readline())
````

````
# 파일에서 줄을 읽을 때, 파일 객체에 루핑 가능
#  list(f) 혹은 f.readline() 사용 가능
with open('7-input.txt') as f:
    for line in f:
        print(line, end='')
````

### f.write( string )
string의 내용을 파일에 쓰고, 출력된 문자의 개수를 리턴

````
with open('7-output.txt', 'w') as f:
    f.write('This is a test\n')
````

````
# 다른 타입의 객체는 변환이 필요
value = ('the answer', 42)  # tuple
s = str(value)            # convert tuple to string
with open('7-output-2.txt', 'w') as f:
    f.write(s)
````

### f.tell()
* 파일의 현재 위치를 가리키는 정수를 리턴
*텍스트 모드의 경우 불투명한 숫자이고,
* 바이너리 모드의 경우 처음부터 바이트 수로 표현

(예제 없음)

### f.seek( offset, whence )
* 객체의 위치를 바꿀 때 사용
* 위치는 기준점에서 offset을 더해서 계산,
* 기준점은 whence인자로 선택.
  * 0: 파일 처음 (기본값), 1: 현재 파일 위치, 2: 파일 끝

````
# 7-input-2.txt instead of workfile
f = open('7-input-2.txt', 'rb+')
f.write(b'0123456789abcdef')
````

````
# Go to the 6th byte 
f.seek(5)
f.read(1)
````

````
# The 3rd before the end
f.seek(-3,2)
f.read(1)
````

````
f.close()
````

### 7.2.2. json 으로 구조적인 데이터를 저장하기
* https://docs.python.org/ko/3.9/tutorial/inputoutput.html#saving-structured-data-with-json

JSON (Java Script Object Notation)
* JSON형식은 데이터 교환을 위해 널리 사용됨.
* 프로그램 간의 연동을 위한 좋은 선택.
#
Q: 왜 JSON을 쓰는가?
A: 까다로운 파싱/직렬화 작업이 간편해집니다.
R: 파일에 읽고 쓸 때 read()가 문자열을 리턴하므로,
    문자열은 쉽지만
    숫자는 타입변환을 위해 int() 등을 사용해야 합니다.
      예: 문자열 '123'을 숫자 123으로 변환
    중첩된 리스트, 딕셔너리처럼 복잡한 데이터를 저장할 때,
    수잡업 파싱 및 직렬화가 까다로울 수 있습니다.
#
json 모듈
    직렬화 (Serializing)   파이썬 데이터 계층 -> 문자열 표현으로 변환
  역질렬화 (Deserializing)                   <-

### dumps()

````
import json
json.dumps([1,'simple','list'])
````

### dump()
객체를 텍스트 파일로 직렬화


````
# x라는 객체가 있을 때
#   예: table = {'Sjoerd': 4127, 'Jack': 4098, 'Dcab': 7678}
# f가 쓰기 위해 열린 텍스트 파일이면
x = table
f = open('7-test-dump.txt', 'w')

# 객체를 텍스트 파일로 직렬화
json.dump(x, f)

f.close()
````
