```
# 4.4. 루프의 break와 continue문, 그리고 else절
#   https://docs.python.org/ko/3.9/tutorial/controlflow.html#break-and-continue-statements-and-else-clauses-on-loops
# review
# revisit
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'equals', x, '*', n//x)
            break
    else:
        # loop fell through without finding a factor
        print(n, 'is a prime number')
```

위의 블록은 올바른 코드입니다. 자세히 들여다보면 else절은 if문이 아닌 for루프에 속합니다. 

* 루프문은 else절을 가질 수 있습니다.
* 루프가 iterable의 소진이나, 조건이 거짓이 되서 종료할 때 실행됩니다.
* 하지만 루프가 break문으로 종료할 때는 실행되지 않습니다.

* 루프와 함께 사용될 때, else절은 if문보다는 try문의 else절과 비슷합니다.
  * try문의 else절은 예외가 발생하지 않을 때 실행됩니다.
    * try문에 대해선 [8.3. 예외 처리하기](https://docs.python.org/ko/3.9/tutorial/errors.html#tut-handling)를 참고하세요.
  * 루프의 else절은 break가 발생하지 않을 때 실행됩니다.

else절이 if문에 속하게 되면 아래와 같이 출력됩니다.

```
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'equals', x, '*', n//x)
            break
        else:
            # loop fell through without finding a factor
            print(n, 'is a prime number')
```

```
# continue문은 C와 같아서 루프의 다음 iteration에서 계속하도록 만듭니다.

for num in range(2, 10):
    if num % 2 == 0:
        print('Found an even number', num)
        continue
    print('Found an odd number', num)
```
