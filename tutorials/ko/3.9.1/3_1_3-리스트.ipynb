```
# 3.1.3. 리스트(list)
#   리스트는 가변(mutable)이므로 내용을 변경할 수 있습니다.
#   https://docs.python.org/ko/3.9/glossary.html#term-mutable
#   참고: 문자열 리터럴은 불변입니다.

squares = [1, 4, 9, 16, 25]
squares
```

```
# indexingUntitled0
squares[0]
```

```
squares[-1]
```

```
squares[-3:]
```

```
# [:]는 얕은 복사본을 리턴합니다.
squares[:]
```

```
squares + [36, 49, 64, 81, 100]
```

```
# 리스트는 가변(mutable)으로 내용을 변경할 수 있습니다.
cubes = [1, 8, 27, 65, 125]
```

```
4 ** 3
```

```
cubes[3] = 64
cubes
```

```
# append()
#   리스트의 끝에 새 항목을 추가합니다.

cubes.append(216)
cubes.append(7 ** 3)
cubes
```

```
# 슬라이스에 대입하는 것도 가능합니다.
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
letters
```

```
letters[2:5] = ['C', 'D', 'E']
letters
```

```
letters[2:5] = []
letters
```

```
# empty list
letters[:] = []
letters
```

```
# Q: letter = []를 쓰면 어떻게 될까?
letters = [1, 2]
letters
```

```
letters = []
letters
```

```
# len()
letters = ['a', 'b', 'c', 'd']
len(letters)
```
