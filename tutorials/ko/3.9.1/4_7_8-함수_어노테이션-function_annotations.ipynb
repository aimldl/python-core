# 4.7.8. 함수 어노테이션
* [4.7.8. 함수 어노테이션](https://docs.python.org/ko/3.9/tutorial/controlflow.html#function-annotations)
* [4.7.8. Function Annotations](https://docs.python.org/3.9/tutorial/controlflow.html#function-annotations)

사용자 정의 함수가 사용하는 타입들에 대한 메타 정보입니다.
* 매개변수 어노테이션 `:`를 이용
* 리턴 값 어노테이션 `->`를 이용 
* 함수의 __annotations__ attribute에 딕셔너리로 저장됩니다.
* 함수의 다른 부분에 아무런 영향을 미치지 않습니다.

자세한 내용은
* [PEP 3107 -- Function Annotations](https://www.python.org/dev/peps/pep-3107/)
* [PEP 484 -- Type Hints](https://www.python.org/dev/peps/pep-0484/)

를 참고하세요.


```
def f(ham: str, eggs: str = 'eggs') -> str:
    print('Annotations:', f.__annotations__)
    print('Arguments:', ham, eggs)
    return ham + ' and ' + eggs
```
