# 4.8. 코딩 스타일
* [4.8. 막간극: 코딩 스타일](https://docs.python.org/ko/3.9/tutorial/controlflow.html#intermezzo-coding-style)
* [4.8. Intermezzo: Coding Style](https://docs.python.org/3.9/tutorial/controlflow.html#intermezzo-coding-style)

파이썬 스타일 가이드로 PEP 8를 참고하세요.
* [PEP 8 -- Style Guide for Python Code](https://www.python.org/dev/peps/pep-0008/)

몇 가지 중요한 부분의 발췌

들어쓰기
* 4개의 스페이스를 씁니다.
* tab을 쓰지 마세요.
* 이유: 작은 들여쓰기와 큰 들여쓰기 사이의 좋은 절충점입니다.

줄넘김
* 79자를 넘기지 않습니다.
* 이유: 
  * 작은 화면을 가진 사용자를 돕고
  * 큰 화면에서도 여러 코드 파일을 나란히 볼 수 있습니다.

빈 줄
* 코드 블록
  * 함수
  * 클래스
  * 함수 내의 큰 코드 블록
* 을 분리하기 위해 빈 줄을 씁니다.
  
스페이스
* a = f(1, 2) + g(3, 4)
  * 연산자들 주변에 스페이스를 넣습니다.
  * 콤마 뒤에 스페이스를 넣습니다.
  * 괄호 안쪽에는 스페이스를 넣지 마세요.

주석 (Comment)
* 가능하다면 다른 줄에 넣습니다.

독스트링 (docstrings) 혹은 도큐멘테이션 문자열 (documentation strings)
* 을 사용하세요. 

이름
* 클래스
  * class UpperCamelCase:
* 함수와 메서드
  * def lowercase_with_underscores():
  * self: 첫 번째 메서드 인자의 이름으로 항상 self를 사용하세요

국제적 환경에서의 사용
* UTF-8 또는 ASCII를 사용하세요.
* 특별한 인코딩을 사용하지 마세요.
* 식별자에 ASCII 이외의 문자를 사용하지 마세요.
  * 다른 언어를 사용하는 사람이 코드를 읽거나 유지할 약간의 가능성이 있더라도...
