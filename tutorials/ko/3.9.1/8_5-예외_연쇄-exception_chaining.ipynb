# `raise ... from ...`
`from`을 쓰면 연쇄적인 예외를 만들 수 있습니다.

* 예외를 변환할 때 유용할 수 있습니다.

```
# revisit
# exc must be exception instance or None
raise RuntimeError from exc
```

except 혹은 finally 섹션에서 예외가 발생하면, 예외 연쇄가 자동으로 일어납니다.

```
# revisit
def func():
    raise IOError

try:
    func()
except IOError as exc:
    raise RuntimeError('Failed to open database') from exc
```

## raise ... from None
`from None` 관용구를 사용하면 예외 연쇄를 비활성화 할 수 있습니다.


```
try:
    open('database.sqlite')
except IOError:
    raise RuntimeError from None
```

* `from None`이 쓰인 위의 경우 첫 번째 예외가 출력되지 않습니다.
* 쓰이지 않은 아래의 경우 첫 번째 예외 다음에 (연쇄적으로) 두 번째 예외도 출력됩니다.

```
try:
    open('database.sqlite')
except IOError:
    raise RuntimeError('What happends without from None?')
```
