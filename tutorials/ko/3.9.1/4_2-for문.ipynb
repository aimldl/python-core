```
# 4.2. for문
#   파이썬의 for문은 C 혹은 파스칼과 약간 다릅니다.
#   파이썬의 for문은
#     임의의 시퀀스 (예: 리스트, 문자열)의 항목들을 순서대로 이터레이션합니다.

words = ['cat', 'window', 'defenestrate']
for w in words:
    print(w, len(w))
```
