````
# + 연산자 -> 문자열을 이어 붙이고
# * 연산자 -> 문자열을 반복
3 * 'un' + 'ium'
````

````
# 두 개 이상의 문자열 리터럴은 자동으로 이어 붙여짐
#   이 기능은 긴 문자열을 쪼갤 때 유용합니다.
#   이 기능은 변수나 표현식에는 해당되지 않습니다. 즉, 변수 '문자열 리터럴'은 자동으로 이어 붙여지지 않습니다.
'Py' 'thon'
````

````
text = ('Put several strings within parentheses'
        'to have them joined together.')
text
````

````
# 변수 '문자열 리터럴'은 자동으로 이어 붙여지지 않습니다.
prefix = 'Py'
prefix'thon'
````

````
('un' * 3) 'ium'
````

````
# 변수 + 문자열 리터럴을 붙일려면 +를 사용합니다.
prefix = 'Py'
prefix + 'thon'
````

````
# 문자열의 인덱스 (서브 스크립트)
# +---+---+---+---+---+---+
#  | P | y | t | h | o | n |
#  +---+---+---+---+---+---+
#  0   1   2   3   4   5   6
# -6  -5  -4  -3  -2  -1
word = 'Python'
word[0]
````

````
word[5]
````

````
# 음의 문자열 인덱스
#   -0과 0은 같으므로 음의 인덱스는 -1에서 시작합니다.
word[-1]
````

````
word[-2]
````

````
word[-6]
````

````
# 인덱스 슬라이싱 (slicing)
#   시작 인덱스는 포함, 종료 위치 문자는 포함되지 않습니다.
word[0:2]
````

````
word[2:5]
````

````
# 종료 인덱스의 문자가 포함되지 않기 때문에
# s[:i] + s[i:] = s 이 됩니다. 
word[:2] + word[2:]
````

````
word[:4] + word[4:]
````

````
word[:2]
````

````
word[4:]
````

````
word[-2:]
````

````
# 에러: 너무 큰 값을 인덱스로 사용할 경우
word[42]
````

````
# 슬라이싱할 때는 범위를 벗어나도 에러 발생 없이 부드럽게 처리됩니다.
word[4:42]
````

````
word[42:]
````

````
# 파이썬 문자열은 불변 (immutable)이므로 내용을 변경할 수 없습니다.
#   문자열 인덱스로 참조한 위치에 대입을 하려고 하면 에러가 발생합니다.
#   https://docs.python.org/ko/3.9/glossary.html#term-immutable
word[0] = 'J'
````

````
word[2] = 'py'
````

````
# 다른 문자열이 필요하면 새로 만들어야 합니다.\
'J' + word[1:]
````

````
word[:2] + 'py'
````
