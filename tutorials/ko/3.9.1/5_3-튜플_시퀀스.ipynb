## 시퀀스 자료형
시퀀스 자료형은 많은 특성을 공유합니다. 예를 들어
  * 인덱싱
  * 슬라이싱 연산

* 리스트 (list)
* 문자열 (string)

에 더해 
* 튜플 (tuple)

도 시퀀스 자료형 입니다.

## 튜플 (tuple)
* 튜플은 쉼표 `,`로 여러 값을 구분합니다.
* 튜플을 입력할 때
  * 괄호를 쓸 수도
  * 안 쓸 수도 있습니다.
* 튜플을 출력할 때
  * 항상 괄호로 둘러쌓입니다.

|           | 튜플 (Tuple)                         | 리스트 (List)                 |
|-----------|--------------------------------------|---------------------------|
| 가변성    | 불변 (immutable)                     | 가변 (mutable)            |
| 구성 요소 | 보통 이질적인 요소들의 시퀀스를 포함 | 보통 등질적인 요소를 포함 |
| 접근 방식 | 인덱싱, 언패킹                       | 이터레이션                |




```
t = 12345, 54321, 'hello!'
t[0]
```

```
t
```

```
# tuples may be nested
u = t, (1, 2, 3, 4, 5)
u
```

```
# tuples are immutable
t[0] = 88888
```

```
# tuples can contain mutable objects
v = ([1, 2, 3], [3, 2, 1])
v
```

## 특별한 튜플
* 빈 튜플
* 하나의 항목을 가지는 튜플

를 수용하기 위한 문법은 추가적인 예외사항을 가지고 있습니다.

하나의 항목으로 구성된 튜플은 값 뒤에 쉼표를 붙입니다.

```
empty = ()
singleton = 'hello',  # 쉼표가 있음에 유의

len(empty)
```

```
len(singleton)
```

`singleton`의 마지막에 쉼표 `,`가 있어서 튜플이 됩니다.

```
singleton
```

```
type(singleton)
```

만약 쉼표 `,`가 없으면, 길이가 5인 리스트가 됩니다.

```
singleton_without_comma = 'hello'
len(singleton_without_comma)
```

```
singleton_without_comma
```

```
type(singleton_without_comma)
```

## 튜플의 언패킹
`t`는 쉼표 `,`를 써서 튜플로 패킹되었습니다.

```
t
```
