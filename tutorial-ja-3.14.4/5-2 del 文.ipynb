## [5.2. del 文](https://docs.python.org/ja/3.14/tutorial/datastructures.html#the-del-statement)

リストから要素を削除する際、値を指定する代わりにインデックスを指定する方法があります。それが del 文です。これは pop() メソッドと違い、値を返しません。 del 文はリストからスライスを除去したり、リスト全体を削除することもできます(以前はスライスに空のリストを代入して行っていました)。例えば以下のようにします:

```python
a =[-1, 1, 66.25, 333, 333, 1234.5]
del a[0]
a
```

```python
del a[2:4]
a
```

```python
del a[:]
a
```

del は変数全体の削除にも使えます:

```python
del a
```

この文の後で名前 a を参照すると、(別の値を a に代入するまで) エラーになります。 

```python
a
```

del の別の用途についてはまた後で取り上げます。
