## [5.4. 集合型 sets](https://docs.python.org/ja/3.14/tutorial/datastructures.html#sets)

Python には、 sets を扱うためのデータ型もあります。

集合とは、重複する要素をもたない、順序づけられていない要素の集まりです。
簡単なデモンストレーションを示します:

```python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(basket)
```

```python
'orange' in basket
```

```python
'crabgrass' in basket
```

### Demonstrate set operations on unique letters from two words

中括弧、または set() 関数は set を生成するために使用することができます。

```python
a = set('abracadabra')
b = set('alacazam')
```

注: 空集合を作成するためには ({} ではなく) set() を使用しなければなりません 。
*  {}は空の辞書を作成します。辞書は次のセクションで議論するデータ構造です。

```python
a
```

```python
b
```

Set オブジェクトは数学的な演算もサポートしています。
  * 和 (union)、
  * 積 (intersection)、
  * 差 (difference)、
  * 対称差 (symmetric difference)

### 和 (union)
aまたはbまたは両方に存在する文字

```python
a | b
```

### 積 (intersection)
aまたはbまたは両方に存在する文字

```python
a & b
```

### 差 (difference)
aのユニークな文字

```python
a - b
```

### 対称差 (symmetric difference)
aまたはbにあるが、両方にはない文字

```python
a ^ b
```

Because sets are unordered, iterating over them or printing them can produce the elements in a different order than you expect.

Similarly to [list comprehensions](https://docs.python.org/ja/3.14/tutorial/datastructures.html#tut-listcomps), set comprehensions are also supported:

```python
a = {x for x in 'abracadabra' if x not in 'abc'}
a
```
