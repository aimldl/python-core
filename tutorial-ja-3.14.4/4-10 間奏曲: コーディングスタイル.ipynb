## [4.10. 間奏曲: コーディングスタイル](https://docs.python.org/ja/3.14/tutorial/controlflow.html#intermezzo-coding-style)
Python には、ほとんどのプロジェクトが守っているスタイルガイドとして PEP 8 があります。
* それは非常に読み易く目に優しいコーディングスタイルを推奨しています。
* 全ての Python 開発者はある時点でそれを読むべきです。
* ここに最も重要な点を抜き出しておきます

<img src="images/PEP8-Style-Guide-for-Python-Code.png">

インデント = 空白 4 つ
* インデントには空白 4 つを使い、タブは使わないこと。
* 空白 4 つは (深くネストできる) 小さいインデントと (読み易い) 大きいインデントのちょうど中間に当たります。タブは混乱させるので、使わずにおくのが良いです。

空白
* a = f(1, 2) + g(3, 4)
  * 演算子の前後とコンマの後には空白を入れ、
  * 括弧類のすぐ内側には空白を入れないこと: 

可能なら、コメントは行に独立で書くこと。

クラスや関数 = { 区切りに空行を使う, クラス名: UpperCamelCase, 関数名/メソッド名: lowercase_with_underscores, self: メソッドの第 1 引数の名前} 
* 関数やクラスや関数内の大きめのコードブロックの区切りに空行を使うこと。
* クラスや関数に一貫性のある名前を付けること。慣習では 
  * UpperCamelCase をクラス名に使い、 
  * lowercase_with_underscores を関数名やメソッド名に使います。
  * 常に self をメソッドの第 1 引数の名前 (クラスやメソッドについては クラス初見 を見よ) として使うこと。

docstring を使うこと。

幅 <= 79 文字
* ソースコードの幅が 79 文字を越えないように行を折り返すこと。
* こうすることで
  * 小さいディスプレイを使っているユーザも読み易くなり、
  * 大きなディスプレイではソースコードファイルを並べることもできるようになります。

あなたのコードを世界中で使ってもらうつもりなら、風変りなエンコーディングは使わないこと。
* どんな場合でも、
  * Python のデフォルト UTF-8 またはプレーン ASCII が最も上手くいきます。
* 同様に、ほんの少しでも他の言語を話す人がコードを読んだりメンテナンスする可能性があるのであれば、
  * 非 ASCII 文字も識別子に使うべきではありません。
