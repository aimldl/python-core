## [4.8. 関数を定義する](https://docs.python.org/ja/3.14/tutorial/controlflow.html#defining-functions)


def は関数の 定義 (definition) を導くキーワードです。 
* def の後には、関数名と仮引数を丸括弧で囲んだリストを続けなければなりません。
* 関数の実体を構成する実行文は次の行から始め、インデントされていなければなりません。

フィボナッチ数列 (Fibonacci series) を任意の上限値まで書き出すような関数を作成できます:

```python
def fib(n):
    """nまでのフィボナッチ数をprintする。"""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
```

```python
# 定義した関数を呼び出す:
fib(100)
```

関数の定義を行うと、関数名は関数オブジェクトとともに現在のシンボルテーブル内に取り入れられます。インタープリタはその名前が指すオブジェクトをユーザ定義関数 (user-defined function) として認識します。

```python
fib
```

他の名前も同じ関数オブジェクトを指すことができ、またその関数にアクセスするために使用することができます:

```python
f = fib
f(100)
```

技術的に言えば、実際には return 文を持たない関数も

```python
fib(0)
```

ややつまらない値ですが値を返しています。

```python
print(fib(0))
```

この値は None と呼ばれます (これは組み込みの名前です)。 None だけを書き出そうとすると、インタプリタは通常出力を抑制します。本当に出力したいのなら、以下のように print() を使うと見ることができます:

フィボナッチ数列の数からなるリストを出力する代わりに、値を返すような関数を書くのは簡単です:

```python
def fib2(n):
    """nまでのフィボナッチ数を含むリストを返す。"""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
        
    return result
```

* この例での append() は result = result + [a] と等価ですが、より効率的です。
* return 文では、関数から一つ値を返します。 
    * return の引数となる式がない場合、 None が返ります。
    * 関数が終了したときにも None が返ります。

```python
f100 = fib2(100)
f100
```

## [4.9. 関数定義についてもう少し](https://docs.python.org/ja/3.14/tutorial/controlflow.html#more-on-defining-functions)

### [4.9.1. デフォルトの引数値](https://docs.python.org/ja/3.14/tutorial/controlflow.html#default-argument-values)

```python
def ask_ok( prompt, retries=4, reminder='再試行してください!'):
    while True:
        reply = input(prompt)
        if reply in ('y', 'ye', 'yes'):
            return True
        if reply in {'n','no','nop','nope'}:
            return False
        retries = retries -1
        if retries < 0:
            raise ValueError('無効なユーザーの入力')
        print(reminder)
```

この関数はいくつかの方法で呼び出せます:

* 必須の引数のみ与える: ask_ok('Do you really want to quit?')
* 一つのオプション引数を与える: ask_ok('OK to overwrite the file?', 2)
* 全ての引数を与える: ask_ok('OK to overwrite the file?', 2, 'Come on, only yes or no!')

### デフォルト値

関数が定義された時点で、関数を 定義している 側のスコープ (scope) で評価されるので

```python
i = 5

def f(arg=i):
    print(arg)
```

```python
f()
```

```python
i = 6
f()
```

重要な警告: 
* デフォルト値は 1 度だけしか評価されません。
* デフォルト値がリストや辞書のような変更可能なオブジェクトの時にはその影響がでます。

例えば以下の関数は、後に続く関数呼び出しで関数に渡されている引数を累積します:

```python
def f(a, L=[]):
    L.append(a)
    return L
```

```python
print( f(1) )
print( f(2) )
print( f(3) )
```

後続の関数呼び出しでデフォルト値を共有したくなければ、代わりに以下のように関数を書くことができます:

```python
def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
```

```python
print( f(1) )
print( f(2) )
print( f(3) )
```
