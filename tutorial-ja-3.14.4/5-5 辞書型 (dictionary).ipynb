## [5.5. 辞書型 (dictionary)](https://docs.python.org/ja/3.14/tutorial/datastructures.html#dictionaries)

もう一つ、有用な型が Python に組み込まれています。それは 辞書 (dictionary) (Mapping Types --- dict を参照)です。辞書は他の言語にも "連想記憶 (associated memory)" や "連想配列 (associative array)" という名前で存在することがあります。

辞書は 
* キー(key): 値(value) のペア
の集合であり、キーが (辞書の中で)一意でなければならない、と考えるとよいでしょう。

カンマで区切られた key: value のペアを波括弧ペアの間に入れると、辞書の初期値となる key: value が追加されます;

以下に、辞書を使った簡単な例を示します:

```python
tel = {'jack': 4098, 'sape': 4139}
```

この表現方法は出力時に辞書が書き出されるのと同じ方法です。

```python
tel
```

ある範囲の数でインデックス化されているシーケンスと異なり、辞書は キー (key) でインデックス化されています。このキーは何らかの変更不能な型になります。

キーにできる
* 文字列、数値は常にキーにすることができます。
* タプルは、文字列、数値、その他のタプルのみを含む場合はキーにすることができます。

キーにできない
* 直接、あるいは間接的に変更可能なオブジェクトを含むタプルはキーにできません。
* リストをキーとして使うことはできません。

これは、リストにスライスやインデックス指定の代入を行ったり、 append() や extend() のようなメソッドを使うと、インプレースで変更することができるためです。

辞書での主な操作は、ある値を何らかのキーを付けて記憶することと、

```python
tel['guido'] = 4127
tel
```

キーを指定して値を取り出すことです。 

```python
tel['jack']
```

すでに使われているキーを使って値を記憶すると、以前そのキーに関連づけられていた値は忘れ去られてしまいます。

```python
tel['sape'] = 1234
tel
```

del で key: value のペアを削除することもできます。

```python
del tel['sape']
```

```python
tel
```

存在しないキーの値をインデックス指定(d[key])で取り出すと KeyError が発生します。

```python
tel['irv']
```

存在しない可能性のあるキーへのアクセス時にこのエラーを回避するには、代わりに get() メソッドを使用します。

```python
print( tel.get('irv') )
```

このメソッドは辞書にキーが存在しない場合に None (または指定されたデフォルト値)を返します。

```python
tel['irv'] = 4127
tel
```

### list(d)
辞書オブジェクトに対し list(d) を実行すると、辞書で使われている全てのキーからなるリストをキーが挿入された順番で返します。


```python
list(tel)
```

### sorted(d)
ソートされたリストが欲しい場合は、代わりに sorted(d) を使ってください。

```python
sorted(tel)
```

### in キーワード
ある単一のキーが辞書にあるかどうか調べるには、 in キーワードを使います。

```python
'guido' in tel
```

```python
'jack' in tel
```

```python
'jack' not in tel
```

### コンストラクタ (constructor)
dict() コンストラクタは、キーと値のペアのタプルを含むリストから辞書を生成します:

```python
dict( [('sape', 4139), ('guido', 4127), ('jack', 4098)] )
```

キーが単純な文字列の場合、キーワード引数を使って定義する方が単純な場合もあります:

```python
dict( sape=4139, guido=4127, jack=4098 )
```

波括弧 (brace) のペア: {} は空の辞書を生成します。

```python
empty_dict = {}
empty_dict
```

```python
empty_dict2 = dict()
empty_dict2
```

さらに、辞書内包表現を使って、任意のキーと値のペアから辞書を作れます:

```python
{x: x**2 for x in (2, 4, 6) }
```
