```
# revisit
# 피보나치 수열(Fibonacci series)의 앞부분을 계산
#   https://ko.wikipedia.org/wiki/%ED%94%BC%EB%B3%B4%EB%82%98%EC%B9%98_%EC%88%98
#   https://en.wikipedia.org/wiki/Fibonacci_number

a, b = 0, 1
while a < 10:
    print(a)
    a, b = b, a+b
```

```
i = 256 * 256
print('The value of i is', i)
```

```
# 키워드 인자 end
#   출력 끝에 포함되는 개행문자를 제거하거나
#   출력을 다른 문자열로 끝나게 하고 싶을 때 사용

a, b = 0, 1
while a < 1000:
    print(a, end=',')
    a, b = b, a+b
```
