```
# 4.3. range() 함수
#   숫자들의 sequence로 iterate할 필요가 있을 때 편리합니다.
```

```
# 끝값인 5는 포함되지 않습니다.
for i in range(5):
    print(i)
```

range함수의 인자가 아래와 같을 때 기대되는 값입니다.

| range function        | 기대하는 값   |
|-----------------------|---------------|
| range(5, 10)          | 5, 6, 7, 8, 9 |
| range(0, 10, 3)       | 0, 3, 6, 9    |
| range(-10, -100, -30) | -10, -40, -70 |

```
# sequence의 index로 iterate하려면
#  range()와 len()을 결합할 수 있습니다.
a = ['Mary', 'had', 'a', 'little', 'lamb']
for i in range(len(a)):
    print(i, a[i])
```

```
# 사실 이럴 땐 enumerate()함수는 쓰는 것이 편합니다.
for i, value in enumerate(a):
    print(i, value)
```

```
# 범위는 값으로 인쇄되지 않습니다.
range(5, 10)
```

```
# 위의 객체는 iterable입니다.
#   https://docs.python.org/ko/3.9/glossary.html#term-iterable
type(range(5, 10))
```

```
# range(4)는 iterable입니다. sum은 iterable을 입력으로 취할 수 있습니다.
sum(range(4))
```
