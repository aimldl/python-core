## 7.1.1. 포맷 문자열 리터럴 (간단히 f-문자열)
*  https://docs.python.org/ko/3.9/tutorial/inputoutput.html#tut-f-strings

접두어로 f 혹은 F를 붙이고 표현식을 작성합니다.

````
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
# : 뒤의 정수는 최소 문자폭
````

````
table = {'Sjoerd': 4127, 'Jack': 4098, 'Dcab': 7678}
for name, phone in table.items():
    print(f'{name:10} -> {phone:10d}')
````

````
# 수정자 Method
# !a    ascii()
# !s    str()
# !r    repr()
animals = 'eels'
print(f'My hovercraft is full of {animals}.')
````

````
print(f'My hovercraft is full of {animals!r}')
````

## 7.1.2. 문자열 format() 메서드
* https://docs.python.org/ko/3.9/tutorial/inputoutput.html#the-string-format-method

str.format() 메서드의 기본 사용법
* [포맷 문자열 문법](https://docs.python.org/ko/3.9/library/string.html#formatstrings)
* [str.format(*args, **kwargs)](https://docs.python.org/ko/3.9/library/stdtypes.html#str.format)

````
print('We are the {} who say "{}!"'.format('knights', 'Ni'))
````

````
# 위치로 지정
print('{0} and {1}'.format('spam', 'eggs'))
print('{1} and {0}'.format('spam', 'eggs'))
````

````
# 변수로 지정
print('This {food} is {adjective}.'.format(food='spam', adjective='absolutely horrible'))
````

````
# 위치와 변수로 지정
print('The story of {0}, {1}, and {other}.'.format('Bill', 'Manfred', other='Georg'))
````

````
# ** 표기법을 사용. 딕셔너리의 key를 인자로 전달
table = {'Sjoerd': 4127, 'Jack': 4098, 'Dcab': 8637678}
print('Jack: {Jack:d}; Sjoerd: {Sjoerd:d}; Dcab: {Dcab:d}'.format(**table))

# 이 방법은 모든 지역 변수를 담은 딕셔너리를 돌려주는 내장함수 vars()와 함께 사용하면 특히 유용함
````

````
for x in range(1, 11):
    print('{0:2d} {1:3d} {2:4d}'.format(x, x*x, x*x*x))
````

## 7.1.3. 수동 문자열 포매팅
위와 동일한 출력을 수동으로 포매팅했습니다.

````
for x in range(1,11):
    print(repr(x).rjust(2), repr(x*x).rjust(3), end=' ')
    print(repr(x*x*x).rjust(4))
# 주의: 각 칼럼 사이에 스페이스 하나가 추가되었음  
````

````
# str.zfill() 숫자 문자열의 왼쪽에 0을 채웁니다.
#   플러스, 마이너스 부호도 이해합니다.
'12'.zfill(5)
````

````
'3.14159265359'.zfill(5)
````

````
# dumps()
import json
json.dumps([1,'simple','list'])
````

````
# dump()
#   객체를 텍스트 파일로 직렬화

# x라는 객체가 있을 때
#   예: table = {'Sjoerd': 4127, 'Jack': 4098, 'Dcab': 7678}
# f가 쓰기 위해 열린 텍스트 파일이면
x = table
f = open('7-test-dump.txt', 'w')

# 객체를 텍스트 파일로 직렬화
json.dump(x, f)

f.close()
````

````
# 객체를 역직렬화
f = open('7-test-dump.txt', 'r')
y = json.load(f)
y
````
