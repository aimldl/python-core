## 사전 지식: 정렬 (sorting)
* [정렬 HOW TO](https://docs.python.org/ko/3/howto/sorting.html)
  * [list.sort()](https://docs.python.org/ko/3/library/stdtypes.html#list.sort)

> key 는 인자 하나를 받아들이는 함수를 지정하는데, 각 리스트 요소에서 비교 키를 추출하는 데 사용됩니다 (예들 들어, key=str.lower). 리스트의 각 항목에 해당하는 키는 한 번만 계산된 후 전체 정렬 프로세스에 사용됩니다. 기본값 None 은 리스트 항목들이 별도의 키값을 계산하지 않고 직접 정렬된다는 것을 의미합니다.

* [Sorting HOW TO](https://docs.python.org/3/howto/sorting.html) (영어)
  * [list.sort()](https://docs.python.org/3/library/stdtypes.html#list.sort)

> key specifies a function of one argument that is used to extract a comparison key from each list element (for example, key=str.lower). The key corresponding to each item in the list is calculated once and then used for the entire sorting process. The default value of None means that list items are sorted directly without calculating a separate key value.

````
# list.sort()의 사용 예
list1 = [1, 4, 5, 3, 1, 5, 7, 11]
list1.sort()
list1
````

````
# list.sort()의 예 2
# 문자열 (string)의 정렬
list_of_string = ['z', 'a', 'e', 't', 'q']
list_of_string.sort()
list_of_string
````

````
list_of_string2 = ['za', 'a', 'e', 't', 'q', 'zb', 'zefwsa', 'zfffsef']
list_of_string2.sort()
list_of_string2
````

````
list_of_keys = ['one', 'two', 'three', 'four']
list_of_keys.sort()
list_of_keys
````

````
# 작은 함수를 인자로 전달할 경우

pairs = [(1, 'one'), (2, 'two'), (3, 'three'), (4, 'four')]  # list of tuples
pairs.sort(key=lambda pair: pair[1])
pairs
````

````
for pair in pairs:
    print(pair)
````

````
for pair in pairs:
    print(pair[0])
````

````
for pair in pairs:
    print(pair[1])
````

## 코드 해설
`pairs.sort(key=lambda pair: pair[1])`에서 pair에 리스트의 각 element가 들어갑니다.
 
 즉, 튜플이 순서대로 들어가게 됩니다.
 ```python
   pair = (1, 'one')
   pair = (2, 'two')
   pair = (3, 'three')
   pair = (4, 'four')
```

 이때
 ```
   pair[0]는 각 숫자,
   pair[1]는 각 숫자의 영어표현입니다.
```

```python
 key=lambda pair: pair[1]
```
는 `pair[1]`인 영어를 정렬하는 key로 쓰겠다는 것입니다.

알파벳 순서로 정렬되므로,
* f로 시작하는 four 먼저
* o로 시작하는 one이 다음
* t로 시작하는 three, two가 각각 3번째, 4번째가 됩니다.

pairs전체가 출력되므로 각 투플이 출력됩니다.

## 딕셔너리의 정렬
참고로 위의 정렬 표현식은 리스트에 적용 가능한 것으로, 딕셔너리는 에러가 발생합니다.

````
# 에러 발생
pairs = {1: 'one', 2: 'two', 3: 'three', 4: 'four'}  # dictionary
pairs.sort(key=lambda pair: pair[1])
pairs
````

> AttributeError: 'dict' object has no attribute 'sort'

리스트는 sort라는 attribute가 있지만, 딕셔너리는 없기 때문에 사용할 수 없습니다. 대신

```python
pairs.sort(  -> sorted(pairs.items(),
```
로 변경하면 코드가 동작합니다.

````
# reference: Python | Sort Python Dictionaries by Key or Value, https://www.geeksforgeeks.org/python-sort-python-dictionaries-by-key-or-value/

pairs = {1: 'one', 2: 'two', 3: 'three', 4: 'four'}  # dictionary
sorted(pairs.items(), key=lambda pair: pair[1])
pairs
````
