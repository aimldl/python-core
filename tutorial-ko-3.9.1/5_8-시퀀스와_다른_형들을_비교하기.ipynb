# 5.8. 시퀀스와 다른 형들을 비교하기
* https://docs.python.org/ko/3.9/tutorial/datastructures.html#comparing-sequences-and-other-types

보통
* 시퀀스 객체는 같은 시퀀스 형의 다른 객체와 비교할 수 있습니다.

만약 비교되는 두 항목의 형이 같으면
* 사전식 비교가 순환적으로 수행됩니다.

## 사전식 비교
* 비교는 사전식 순서를 사용합니다.
  * 먼저 첫 두 항목을 비교해서
    * 다르면, 이것이 비교 결과를 결정합니다.
    * 같으면, 다음 두 항목을 비교합니다.
  * 이렇게 한 시퀀스의 끝까지 계속합니다.
  * 두 시퀀스의 모든 항목를 비교한 결과가 같다면, 두 시퀀스는 같은 것으로 취급됩니다.

* 문자열의 사전식 배열은 개별 문자들의 순서를 정하는데
  * 유니코드 코드포인트 숫자를 사용합니다.

## 유의
서로 다른 형의 객체들을 < 나 >로 비교하는 것은
* 그 객체들이 적절한 비교 메서드를 가지고 있을 때만 허락됩니다.
  * 예
    * 서로 다른 숫자 형들은 그 숫자 값으로 비교됩니다.
    * 0과 0.0은 같습니다. 1과 1.0도 같습니다. 와 같이 진행됩니다.
* 적절한 메서드가 없다면, `TypeError`를 일으킵니다.


## 같은 형의 시퀀들을 비교하는 몇 가지 예

```
(1, 2, 3) < (1, 2, 4)
```

```
[1, 2, 3] < [1, 2, 4]
```

```
# revisit
'ABC' < 'C' < 'Pascal' < 'Python'
```

```
(1, 2, 3, 4) < (1, 2, 4)
```

```
# revisit
(1, 2) < (1, 2, -1)
```

```
(1, 2, 3) == (1.0, 2.0, 3.0)
```

```
# revisit
(1, 2, ('aa', 'ab')) < (1, 2, ('abc', 'a'), 4)
```
