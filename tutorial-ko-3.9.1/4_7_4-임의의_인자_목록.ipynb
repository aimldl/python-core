```
# 4.7.4. 임의의 인자 목록
#   https://docs.python.org/ko/3.9/tutorial/controlflow.html#arbitrary-argument-lists
#
#   함수 정의를 할 때 가장 덜 사용되는 옵션입니다.
#   인자의 길이가 바뀔 수 있고, 
#   인자들은 튜플로 묶입니다., 
#
#   보통 매개변수 목록의 마지막에 옵니다.
#   *args 매개변수 뒤에 등장하는 매개변수는 모두 <키워드-전용> 인자로,
#   위치 인자 대신 키워드 인자로만 사용될 수 있습니다.
```

```
def write_multiple_items(file, separator, *args):
    file.write(separator.join(args))
```

```
# review
def concat(*args, sep='/'):
    return sep.join(args)
```

```
concat('earth', 'mars', 'venus')
```

```
concat('earth', 'mars', 'venus', sep='.')
```
