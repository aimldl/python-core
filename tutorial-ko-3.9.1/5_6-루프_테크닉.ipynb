## `items()`: 딕셔너리의 루프
`items()`메서드는 키와 값을 리턴합니다.

```
knights = {'gallahad': 'the pure', 'robin': 'the brave'}

for key, value in knights.items():
    print(key, value)
```

## `enumerate()`: 시퀀스 (리스트, 문자열, 튜플)의 루프
`enumerate()`함수는 인덱스와 해당 값을 리턴합니다.

```
a_list = ['tic', 'tac', 'toe']
for index, value in enumerate(a_list):
    print(index, value)
```

## `zip()`: 두 개 이상의 시퀀스를 동시에 루프
`zip()`함수는 엔트리의 쌍을 리턴합니다.

```
questions = ['name', 'quest', 'favorite color']
answers = ['lancelot', 'the holy grail', 'blue']

for question, answer in zip(questions, answers):
    print('What is your {0}? It is {1}.'.format(question, answer))
```

## `reversed()`: 시퀀스의 역순 루프
`reversed()`함수는 정방향 시퀀스를 역순으로 리턴합니다.

```
for i in reversed(range(1, 10, 2)):
    print(i)
```

## sorted(): 시퀀스를 정렬된 순서로 루프


```
basket = ['apple', 'orange', 'apple', 'pear', 'orange', 'banana'] 

for i in sorted(basket):
    print(i)
```

## `set()` & `sorted()`: 시퀀스의 고유한 요소를 정렬된 순서로 루프
`set()'은 시퀀스의 중복 요소를 제거합니다.

```
basket = ['apple', 'orange', 'apple', 'pear', 'orange', 'banana']

for i in sorted(set(basket)):
    print(i)
```

## 주의: 루프를 돌고 있는 리스트의 변경
* 루프를 돌고 있는 리스트를 변경하지 마세요.
* 새 리스트를 만드는 것이 더 안전하고 간단합니다.

```
# raw_data에 있는 nan를 필터링하는 프로그램

import math

raw_data = [56.2, float('NaN'), 51.7, 55.3, 52.5, float('NaN'), 47.8]
raw_data
```

```
filtered_data = []
for value in raw_data:
    if not math.isnan(value):
        filtered_data.append(value)

filtered_data
```
