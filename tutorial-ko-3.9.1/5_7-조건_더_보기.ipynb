## 비교 연산자
* 특정값이 시퀀스에 있는지/없는지 검사합니다.
  * in
  * not in

* 두 가변 객체가 같은지를 비교합니다.
  * is
  * is not
  * 리스트 같은 가변 객체에서만 의미가 있습니다.

* 논리 연산자를 결합할 수 있습니다.
  * not
  * and
  * or

* 단락 회로 연산자 (short-circuit operator)
  * and와 or는 단락 회로 연산자입니다.
    * 왼쪽에서 오른쪽으로 값이 구해집니다.
    * 결과가 결정되자 마자 값 구하기는 중단됩니다.
    * 예: A = True, B = False, C = True 일 때,
      * `A and B and C`의 경우, C의 값은 구하지 않습니다.

* 우선 순위
  * 산술 연산자 > 비교 연산자 > 논리 연산자
    * 논리 연산자: not > and > or
      * 예: `A and not B or C` = `(A and (not B)) or C`



## 주의: 흔히 하는 실수
* `==`를 사용할 표현식에 `=`를 입력

* 비교의 결과
* 다른 논리 표현식의 결과
* 를 변수에 대입할 수 있습니다.

```
string1, string2, string3 = '', 'Trondheim', 'Hammer Dance'
non_null = string1 or string2 or string3
non_null
```
