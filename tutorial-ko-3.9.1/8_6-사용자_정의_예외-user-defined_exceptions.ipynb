# 8.6. 사용자 정의 예외 (User-defined Exceptions)
* https://docs.python.org/ko/3.9/tutorial/errors.html#user-defined-exceptions

사용자가 예외를 새롭게 정의하면 사용자 정의 예외가 됩니다.
* 사용자 정의 예외를 만들려면 새로운 예외 클래스를 만들면 됩니다.
  * 보통 `Exception`클래스를 직/간접적으로 상속합니다.
  * 예외 클래스를 보통 간단하게 유지합니다.
  * 예외의 이름은 대부분 `Error`로 끝납니다.
    * Error, InputError, TransitionError

* 많은 표준 모듈은 정의된 함수에서 발생가능한 예외를 자체적으로 정의합니다.
  * 더 자세한 정보는 다음을 참고하세요.
    * [9. 클래스](https://docs.python.org/ko/3.9/tutorial/classes.html#tut-classes)
    * [9. Classes](https://docs.python.org/3.9/tutorial/classes.html#tut-classes)

여러가지 에러를 일으키는 모듈을 만들 때
* 흔히 베이크 클래스를 정의한 다음
* 각기 다른 에러 조건 마다 특정한 예외 클래스를 서브 클래스로 만듭니다.


```
class Error(Exception):
    """ Base class for exceptions in this module."""
    pass

class InputError(Error):
    """Exception raise for errors in the input.

    Attributes:
        expression -- input expression in which the error occured
        message -- explanation fo the error
    """

    def __init__(self, expression, message):
    self.expression = expression
    self.message = message

class TransitionError(Error):
    """Raised when an operation attempts a state transition that's not allowed.

    Attributes:
        previous -- state at beginning of transition
        next -- attempted new state
        message -- explanation of why the specific transition is now allowed
    """

    def __init__(self, previous, next, message):
        self.previous = previous
        self.next = next
        self.message = message
```
